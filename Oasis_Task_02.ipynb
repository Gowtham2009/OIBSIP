```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df2[' Date'], df2[' Estimated Unemployment Rate (%)'])
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title('Unemployment Rate Over Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 6))
plt.bar(df2['Region'], df2[' Estimated Unemployment Rate (%)'])
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title('Unemployment Rate Among Region')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(df2.groupby('Area')[' Estimated Employed'].sum(), labels=df2['Area'].unique(), autopct='%1.1f%%', startangle=140)
plt.title('Employment Distribution by Area')
plt.axis('equal')
plt.show()
```

```python
df3 = df2.groupby(['Region', 'Area'])[' Estimated Unemployment Rate (%)'].sum().unstack()

df3.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Unemployment Rate by Area and Region')
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.legend(title='Area')
plt.tight_layout()
plt.show()
```

```python
ap = df2[df2['Region'] == 'Andhra Pradesh']
ap[' Date'] = pd.to_datetime(ap[' Date'])
ap = ap.sort_values(by=' Date')

plt.figure(figsize=(10, 6))
plt.plot(ap[' Date'], ap[' Estimated Unemployment Rate (%)'], marker='o', linestyle='-')
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title('Unemployment Rate in Andhra Pradesh Over Time')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
